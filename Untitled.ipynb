# ASSIGNMENT  

```python
speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "Speed Low"
    elif val == speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"

data["Speed_high_low"] = data["Speed"].apply(set_speed)

HP_mean = data["HP"].mean()

def set_HP(val):
    if val < HP_mean:
        return "HP Low"
    elif val == HP_mean:
        return "HP neutral"
    else:
        return "HP High"

data["high_low_HP"] = data["HP"].apply(set_HP)
```
